# Univariate Analysis

```python
#Age distribution
fig, ax = plt.subplots(figsize=(14, 8))
df['age'].plot(kind='hist', bins=40, color='blue', edgecolor='black', ax=ax)
ax.set_title('Age Distribution', fontsize=15)
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
plt.show()
```

The most represented age group in the dataset is from 20 to 36 years old, with a modest presence of individuals aged between 18 and 20 years as well as those between 36 and 55 years old.

```python
#  Same Race 1= yes, 0=no
df['samerace'] = df['samerace'].map({ 0 : "No", 1 : "Yes"})
```

```python
df["race"] = df["race"].apply(lambda x: 'Black/African American' if x == 1
                                  else 'European/Caucasian-American' if x == 2
                                  else 'Latino/Hispanic American' if x == 3
                                  else 'Asian/Pacific Islander/Asian-Americann' if x == 4
                                  else 'Native American' if x == 5
                                   else 'Other'
                                         )
```

```python
fig = px.histogram(df, x='race', category_orders={'race': df['race'].value_counts().index})
fig.update_layout(
    height=800, width=1000,
    title_text="Distribution of race for all of the participants"
)
fig.show()
```

We see that the most of participants are European/Caucasian-American

```python
plt.figure(figsize=(14, 8))
df['imprace'].value_counts().sort_index().plot(kind='bar', color=plt.cm.Paired(range(10)), edgecolor='black', title='Importance of Same Racial', ylabel='Count', xlabel='Importance race', fontsize=15)
plt.show()
```

The race is not important for the most of people. 

```python
plt.figure(figsize=(14, 8))
df['imprelig'].value_counts().sort_index().plot(kind='bar', color=plt.cm.Paired(range(10)), edgecolor='black', title='Importance of Same Religion', ylabel='Count', xlabel='Importance religion', fontsize=15)
plt.show()
```

The religion is not important too

```python
#What people are looking for?
df["goal"] = df["goal"].apply(lambda x: 'Seemed like a fun night out' if x == 1
                                  else 'To meet new people' if x == 2
                                  else 'To get a date' if x == 3
                                  else 'Looking for a serious relationship' if x == 4
                                  else 'To say I did it' if x == 5
                                   else 'Other'
                                         )
```

```python
#plot with details What people are looking for in a partner
plt.figure(figsize=(14, 8))
df['goal'].value_counts().plot(kind='bar',color=plt.cm.Paired(range(10)), edgecolor='black', title='What people are looking for', ylabel='Count', xlabel='Goal', fontsize=15)
plt.show() 
```

Most of people are looking for fun night out

# Bivariate Analysis

## Distribution of age by gender

```python
age_women=df[df["gender"]==0]['age']
age_men=df[df["gender"]==1]['age']

plt.figure(figsize=(14, 7))


plt.subplot(1, 2, 1)
age_women.plot.hist( bins=40, color='red')
plt.axvline(age_women.median(), color='black', linestyle='dashed', linewidth=2)
plt.text(age_women.median(), plt.gca().get_ylim()[1]*0.9, f'Median Women: {age_women.median():.1f}', color='black', ha='center')

plt.title('Age Distribution of women')
plt.xlabel('Age')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)

df[df["gender"]==1]['age'].plot(kind='hist', bins=40, color='blue')
plt.axvline(age_men.median(), color='black', linestyle='dashed', linewidth=2)
plt.text(age_men.median(), plt.gca().get_ylim()[1]*0.9, f'Median men: {age_men.median():.1f}', color='black', ha='center')
plt.title('Age Distribution of men')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Graphics show that women are younger than men.

## what are people looking for on the opposit sex 

```python
df = df.rename(columns={"attr1_1":"attractive", "sinc1_1":"sincere","intel1_1":"intelligent","fun1_1":"fun","amb1_1":"ambitious","shar1_1":"shared_interests"})
```

```python
attributes = ["attractive", "sincere", "intelligent", "fun", "ambitious", "shared_interests", "gender"]
df_pref = df[attributes].groupby('gender').mean().reset_index()
df_pref=df_pref.melt(id_vars='gender')
plt.figure(figsize=(12, 8))
sns.barplot(data=df_pref, x='variable', y='value', hue='gender',palette={0:'red',1:'blue'})
plt.xlabel('Attributes')
plt.title('Mean Ratings of Attributes by Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Men look for more attractivity and intellegence
* women look for intellegence and sincerity
